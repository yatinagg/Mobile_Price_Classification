## Formatting your Data – making data types compatible with other data types.

```python
df
```

```python
dfm = df.copy()
```

```python
dfm['GPRS'].fillna(np.nan,inplace=True)
dfm['EDGE'].fillna(np.nan,inplace=True)
```

```python
dfm
```

```python
dfm.loc[dfm['GPRS'] == 'Yes', 'GPRS'] = 1
dfm.loc[dfm['GPRS'] == 'No', 'GPRS'] = 0
dfm.loc[dfm['GPRS'] == np.nan, 'GPRS'] = 0
dfm.loc[dfm['EDGE'] == 'Yes', 'EDGE'] = 1
dfm.loc[dfm['EDGE'] == 'No', 'EDGE'] = 0
dfm.loc[dfm['EDGE'] == np.nan, 'EDGE'] = 0
dfm.head()
```
